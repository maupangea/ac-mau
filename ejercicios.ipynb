## Coeficiente de Correlación de Pearson

Sean $X=(X_1,...X_n)$ y $Y=(Y_1,...,Y_n)$ dos variables aleatorias reales, el *coeficiente de correlación de Pearson* de $X$ y $Y$ denotado por $r_{XY}$, se define como:
$$ r_{XY} = \frac{\sum(X_i - \bar{X})(Y_i - \bar{Y})}{\sum(X_i - \bar{X})^2 \sum(Y_i - \bar{Y})^2} $$
